## Creating the columns for the dataframe 

```python
columns_in_original=['avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
considerations=['mean','std','min','1quar','median','3quar','max']
columns_formed=['instance','filename']
for i in columns_in_original:
    for j in considerations:
        columns_formed.append(i+'_'+j)
dataset=pd.DataFrame(columns=columns_formed)
dataset
```

### for a better idea of the below logic, first i am looping through all the files and then i am also looping through the describe of each dataset( i have rectified the missing datasets here) and i have formed individual columns in such a way that looping through the describe and creating a row exactly match the column order above. Finally i am appending the row into the dataframe based on the current index and instance 

```python
directories=[]
for (dirpath, dirnames, filenames) in walk('../../data/AReM/'):
    directories.append(dirpath)
directories=directories[1:]
instance=1

for ii in directories:
    #print(i)
    for dirpath,dirnames,filenames in walk(ii):
        filenames=sorted(filenames)
        for i in filenames:
            #print(dirpath+'/'+i)
            #data=pd.read(dirpath+'/'+i,skiprows=4)
            data=pd.read_csv(dirpath+'/'+i,skiprows=4)
            #print(i,len(data))
            data.drop(data.columns[0],axis=1,inplace=True)
            ##Forming a row
            desc=data.describe()
            row=[instance,dirpath+'/'+i]
            #dataset['instance']=instance
            instance+=1
            for col in desc.columns:
                for val in desc[col][1:]:
                    #print(val)
                    #dataset[col+'_'+val]=desc[col][val]
                    row.append(val)
            dataset.loc[len(dataset.index)]=row
        
dataset
```

### Calculating the standard deviation of the features

```python
std_conf=pd.DataFrame(columns=['feature','std','conf_int_90perc'])
for i in dataset.columns[2:]:
    #print(i)
    row=[i]
    row.append(np.std(dataset[i],ddof=1))
    #print(dataset[i].std(),np.std(dataset[i],ddof=1))
    if(min(dataset[i])==max(dataset[i])):
        row.append('low='+str(min(dataset[i]))+', high='+str(min(dataset[i])))
    else:
        res = bootstrap((dataset[i],), np.std, confidence_level=0.9)
        row.append(str(res.confidence_interval).split('(')[1].split(')')[0])
    #print(row)
    std_conf.loc[len(std_conf.index)]=row
    
std_conf   
```
